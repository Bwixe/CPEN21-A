##For Loop Statement

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
```

##The Break Statement

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
  if x=="Thursday":
   break
```

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  if x =="Thursday":
    break
  print(x)
```

##Looping Through String

```python
for x in "week":
  print(x)
```

##The Range () Function

```python
for x in range(6):
  print(x)

for x in range(2,6):
  print("Example 2",x)
```

##Nested Loops

```python
adjective = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]

for x in adjective:
  for y in fruits:
   print(x,y)
```

##While Loop Statement

```python
i = 1
while i<6:
  print(i)
  i+=1#i+1 Assignment operator 
```

##The Break Statement

```python
i = 1
while i<6:
  print(i)
  if i == 3:
    break
  i+=1 #Assignment operator for addition
```

```python
i = 1
while i<6:
  i+=1 #Assignment operator for addition
  if i==3:
    continue
  else:
    print(i)
```

##The Else Statement

```python
i = 1
while i<6:
  print(i)
  i+=1
else:
  print("no longer less than 6")
```
